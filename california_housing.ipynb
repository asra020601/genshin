## **Given dataset represents housing in California, USA in the year 1990. It contains 20,640 entries with columns such as longitude, latitude, housing median age, total number of rooms, total number of bedrooms, population, number of households, median income, ocean_proximity, median house value.**


---
## **Here the columns : longitude, latitude, median income, median house value are numerical (quantitative)continuous features and the columns : housing median age, total number of rooms, total number of bedrooms, population, number of households are numerical (quantitaive) discrete features whereas ocean_proximity is a categorical (nominal) feature.**


```python
import pandas as pd
import matplotlib.pyplot as plt
#importing pandas and matplotlib and
df = pd.read_csv('housing.csv')
#reading the given dataset
df.info()
```

```python
f = df[['median_income']].mean()
#using mean() method to find out the mean of median income
k = df.median_income
#assigning the given column to a variable 
print("Median income is ",f)
#printing median income
plt.hist(k)
#using matplotlib to plot a histogram of median income
plt.show()
#using matplotlib to show a histogram of median income
```


### Here we observe that the histogram of the median income of the residents in california is a right skewed (non symmetric) distribution. This implies that the mode is bigger than the mean.






```python
a = df.housing_median_age
#assigning the given column to a variable
plt.hist(a,density = 1,color ='green',alpha = 0.7)
#using matplotlib to plot a histogram
plt.show()
#using matplotlib to show the histogram
```

### As we can see the given histogram represents a bimodal distribution. Here the median, mean are nearly equal but mode is vastly different. The mean age of housing in california is 29 whereas 52 is the most recurring age.

```python
import seaborn as sns
sns.scatterplot(x = "median_income", y = "median_house_value", hue = "ocean_proximity", data = df)
#using scatter to plot a scatter plot
plt.show()
#using matplotlib to show the scatter plot
```

### A scatter plot is used here to find the co-relation between two continuous data features. Here a positive co-relation is observed between median house value and median income. This shows that as median house value increases, an increase is also seen in the median income. As the plot shows, houses closer to ocean are more expensive and inland houses are cheaper.

### In the above code, the mean/average of the whole column replaces the null values in the column.

```python
d = df.longitude
#assigning the given column to a variable
e = df.latitude
#assigning the given column to a variable
plt.scatter(e,d)
#using matplotlib to plot a scatter plot
plt.show()
```

### From the above scatter plot we observe that when one value increases the other decreases. This implies negative co-relation between two variables. Latitude and Longitude are negatively co-related.

```python
indexPlace = df[ (df['ocean_proximity'] != "NEAR BAY") ].index
#pandas index is used to select data and store it in a variable
df2 = df.drop(axis = 1, index = indexPlace)
#a new dataset is created to and all the unneccssary values are dropped
df2.info()
```

### In the above code, we delete all the values that aren't necessary and stored in a different dataset. As a result, a drastic change is seen in the number of entries. The columns are still the same but the rows seem significantly decreased. There are 2290 houses near bay houses in california out of 20,640 houses in total.

```python
print(df2["median_income"].mean())
#mean function is used to calculate the mean of the median income
print(df2["median_income"].median())
#median function is used to calculate the median of the median income
```

### The mean and mode are very close to each other, this implies that the data may be symmetric.
